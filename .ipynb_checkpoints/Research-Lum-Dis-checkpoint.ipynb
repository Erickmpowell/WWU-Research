```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#this 'magic' matplotlib command tells iPython to show any figures in this notebook, 
#rather than a separate window or saving them in separate files.
%matplotlib inline

#These commands help python remember how to format my figures so that they look nice
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('pdf', 'png')
plt.rcParams['savefig.dpi'] = 200

plt.rcParams['figure.autolayout'] = False
plt.rcParams['figure.figsize'] = 10, 6
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 20
plt.rcParams['font.size'] = 16
plt.rcParams['lines.linewidth'] = 2.0
plt.rcParams['lines.markersize'] = 8
plt.rcParams['legend.fontsize'] = 14

matplotlib.rcParams['xtick.direction'] = 'out'
matplotlib.rcParams['ytick.direction'] = 'out'
```

```python
coma = np.loadtxt('comaallzad.tab.010.q25mem.tab', skiprows=2, unpack=True)
```

```python
ff, (ax1, ax2) = plt.subplots(2)

#Cutoff Radius for sample, cutoff luminosities for bounding the graphs



n=0
cutoffR=1
cutoffLumL=14
cutoffLumU=18

while coma[10,n]<cutoffR:
    if coma[6,n]<cutoffLumU: 
        ax1.scatter(coma[6,n],coma[2,n],s=10,c='b')
    if coma[6,n]<cutoffLumL and coma[6,n]>12: 
        ax2.scatter(coma[6,n],coma[2,n],s=10,c='b')
    n=n+1
    
ax1.set_title("Redshift for all members with radial cut <1")  
ax2.set_title("Redshift for members with 12 < m$_r$ <14")    
ax1.axvline(x=12,color='k')    
ax1.axvline(x=13,color='k')
ax1.axvline(x=14,color='k')
ax1.axvline(x=15,color='k')
ax1.axvline(x=16,color='k')
ax1.axvline(x=17,color='k')

ax1.set_xlabel("Magnitude in r")
ax1.set_ylabel("CZ")
ax2.set_xlabel("Magnitude in r")
ax2.set_ylabel("CZ")
ff.tight_layout()
```

```python
## taking the coordinates of coma and splitting into low raidus lists for the graph
n=0
lowRcoma=[[],[]]
while coma[10,n]<cutoffR:
    lowRcoma[0].append(coma[0,n])
    lowRcoma[1].append(coma[1,n])
    n=n+1
```

```python
#taking all the data and finding the restframe velocity
#10 bins in this case 

##try 3 bins for better results


radialCut=1
bins=10
binDif=((max(coma[6])-min(coma[6]))/bins)
comaBinnedVelocities= []
vRF=[]
tempBin=[]
vrfTemp=[]
n=0
z=0

##my attempt to split coma into bins that took forever but sucks
totalAverage = np.average(coma[2])
comaVRF=abs((coma[2]-totalAverage))

for z in range(bins):
    tempBin=[]
    vrfTemp=[]
    n=0
    while n<len(coma[6]):            
        if coma[6,n]>=(min(coma[6])+binDif*(z+1)) and coma[6,n]<=(min(coma[6])+binDif*(z+2)):
            tempBin.append(coma[2,n])
            vrfTemp.append(abs((coma[2,n]-totalAverage)))
        n=n+1
    comaBinnedVelocities.append(tempBin)
    vRF.append(vrfTemp)
    
averageVRF=[]
for c in range(10):
    average=np.average(vRF[c])
    averageVRF.append(average)
    
#i want VRF of every element

print(comaVRF[0],coma[2,0],coma[2,0]-totalAverage)
```
